```python
from kafka import KafkaProducer
import json

producer = KafkaProducer(
    bootstrap_servers=['kafka:9093'],
    value_serializer=lambda message: json.dumps(message, default=str).encode("ascii"),
)

future = producer.send('my-topic', b'bonjour ca va')
```

```python
import json
import time
import urllib.request

from kafka import KafkaProducer

API_KEY = "e4df5b72244600359560dc58c04e50681f61eeed"
url = "https://api.jcdecaux.com/vls/v1/stations?apiKey={}".format(API_KEY)

producer = KafkaProducer( bootstrap_servers=['kafka:9093'])

while True:
    response = urllib.request.urlopen(url)
    stations = json.loads(response.read().decode())
    for station in stations:
        producer.send("velib-stations", json.dumps(station).encode())
    print("{} Produced {} station records".format(time.time(), len(stations)))
    time.sleep(1)
```
